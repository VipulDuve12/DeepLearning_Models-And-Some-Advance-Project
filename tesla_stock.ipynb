```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv('/content/TSLA.csv')
```

```python
# making the new variable with next day opening prize.
data['Next_Day'] = data['Close'].shift(-1)
```

```python
data['Target'] = data['Next_Day'] > data['Close']
```

```python
data
```

```python
data['Target'] = data['Target'].astype(int)
```

```python
data
```

```python
data.dropna()
```

```python
predectors = ['Open','High','Low','Close','Volume']

# train test split
train = data.iloc[:-500]
test = data.iloc[-500:]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100,min_samples_split=200,random_state=1)
```

```python
model.fit(train[predectors],train['Target'])
```

```python
from sklearn.metrics import precision_score
```

```python
pred = model.predict(test[predectors])
```

```python
pred = pd.Series(pred,index=test.index)
```

```python
score = precision_score(test['Target'],pred)
```

```python
score
```

```python
graph = pd.concat([test['Target'],pred],axis=1)
```

```python
# orange lines are our predicted value and blue is actual happening....
graph.plot()
```

```python
def prediction(train,test,predectors,model):
  model.fit(train[predectors],train['Target'])
  pred = model.predict(test[predectors])
  pred = pd.Series(pred,index=test.index,name='Predictions')
  graph = pd.concat([test['Target'],pred],axis=1)
  return graph
```

```python
def backtest(data1,model,predectors,start=1500,step=250):
  all_predictions = []
  for i in range(start,data.shape[0],step):
    train = data.iloc[0:i].copy()
    test = data.iloc[i:(i+step)].copy()
    predictions = prediction(train,test,predectors,model)
    all_predictions.append(predictions)
  return pd.concat(all_predictions)
```
